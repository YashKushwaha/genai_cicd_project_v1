```python
import boto3

iam_client = boto3.client('iam')
role_name = "ec2_project_roleAV"

def find_instance_profile_for_role(role_name):
    paginator = iam_client.get_paginator('list_instance_profiles')
    for page in paginator.paginate():
        for profile in page['InstanceProfiles']:
            for role in profile['Roles']:
                if role['RoleName'] == role_name:
                    return profile['InstanceProfileName']
    return None

profile_name = find_instance_profile_for_role(role_name)
if profile_name:
    print(f"Found instance profile: {profile_name}")
else:
    print("No instance profile found for the given role.")
```
